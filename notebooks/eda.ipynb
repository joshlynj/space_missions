```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  
import seaborn as sns
```

```python
launch_df = pd.read_csv("../data/mission_launches.csv")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(launch_df['Price'], bins=20, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Mission_Status', data=launch_df)
plt.title('Mission Status Count')
plt.xlabel('Mission Status')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Organisation', data=launch_df, order=launch_df['Organisation'].value_counts().index)
plt.title('Mission Count by Organisation')
plt.xlabel('Organisation')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```
